```python
movies=df[df["type"]=='Movie'].copy()
movies['duration'] = movies["duration"].str.replace(" min"," ")
movies['duration']=pd.to_numeric(movies['duration'],errors='coerce')
max_dur=movies['duration'].max()
avg_dur=movies['duration'].mean()
print("average time:",avg_dur)
print("longest time:",max_dur)
```

Final Conclusion:

1. Movies are more than TV Shows on Netflix.
2. Drama and Comedy are among the most popular genres.
3. USA produces the most Netflix content.
